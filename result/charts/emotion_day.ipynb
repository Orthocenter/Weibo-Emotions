```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

data = []

with open('emotion_time2', 'r') as f:
    for line in f:
        date, emotion = line.split()
        emotion = float(emotion)
        data.append((date, emotion))

data = sorted(data)
dates, emotions = zip(*data)

plt.plot(emotions)
plt.grid(True)
x = range(0, len(dates), 15)
_ = plt.xticks(x, [dates[i] for i in x], rotation=45)
plt.xlim([0, len(data)])

plt.savefig('emotion_time2', dpi=600)
```
