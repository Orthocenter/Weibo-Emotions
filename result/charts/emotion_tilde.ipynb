```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

data = [307349, 845398, 135452]
```

```python
labels = 'Ambivalent', 'Positive', 'Negative'
colors = ['yellowgreen', 'lightskyblue', 'orange']
explode = [0.1, 0.1, 0.1]

plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%',
        startangle=90, explode=explode, shadow=True)
plt.axis('equal')

plt.savefig('emotion_tilde.png', dpi = 600)
```
